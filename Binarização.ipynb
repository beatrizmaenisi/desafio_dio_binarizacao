```python
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt
```

```python
#Função para carregar e converter para escala em cinza
def carregar_e_converter_para_cinza(caminho_imagem):
        imagem_original = Image.open(caminho_imagem)
        largura, altura = imagem_original.size
        pixels =  imagem_original.load()

        # Criar uma nova imagem em escala de cinza
        imagem = Image.new("L", (largura, altura)) #'L' = escala de cinza

        # Converter a imagem para escala de cinza
        imagem_cinza = imagem.convert("L")

        for y in range(altura):
          for x in range(largura):
            r, g, b = pixels[x,y]
            #Fórmula para converter RGB em cinza
            cinza = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
            imagem_cinza.putpixel((x, y), cinza)

        return imagem_cinza
```

```python
def binarizar_imagem(imagem, limiar):
    largura, altura = imagem_cinza.size
    pixels = imagem_cinza.load()
    imagem_binaria = Image.new("L", (largura, altura))

    for y in range(altura):
        for x in range(largura):
          valor_cinza = pixels[x,y]
          if valor_cinza > limiar:
            imagem_binaria.putpixel((x, y), 255)
          else:
            imagem_binaria.putpixel((x, y), 0)

    return imagem_binaria
```

```python
limiar = 130
imagem_binaria = binarizar_imagem(imagem_cinza, limiar)
```
